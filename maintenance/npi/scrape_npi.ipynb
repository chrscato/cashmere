---convert to parquet

```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

csv_path = r"C:\Users\ChristopherCato\Downloads\NPPES_Data_Dissemination_April_2025\npidata_pfile_20050523-20250413.csv"
parquet_path = r"C:\Users\ChristopherCato\Downloads\NPPES_Data_Dissemination_April_2025\radiology_only.csv"

chunk_size = 100_000

columns_to_keep = [
    # Identification
    "NPI",
    "Entity Type Code",
    "Provider Organization Name (Legal Business Name)",
    
    # Mailing Address (basic location info)
    "Provider First Line Business Mailing Address",
    "Provider Second Line Business Mailing Address",
    "Provider Business Mailing Address City Name",
    "Provider Business Mailing Address State Name",
    "Provider Business Mailing Address Postal Code",
    "Provider Business Mailing Address Telephone Number",
    "Provider Business Mailing Address Fax Number",

    # Practice Location Address (can differ from mailing)
    "Provider First Line Business Practice Location Address",
    "Provider Second Line Business Practice Location Address",
    "Provider Business Practice Location Address City Name",
    "Provider Business Practice Location Address State Name",
    "Provider Business Practice Location Address Postal Code",
    "Provider Business Practice Location Address Telephone Number",
    "Provider Business Practice Location Address Fax Number",

    # Taxonomy Codes + Description + Group (just for 1, can expand to 215 later)
    "Healthcare Provider Taxonomy Code_1",
    "Healthcare Provider Primary Taxonomy Switch_1",
    "Healthcare Provider Taxonomy Group_1",

    # Contact for Org
    "Authorized Official First Name",
    "Authorized Official Last Name",
    "Authorized Official Title or Position",
    "Authorized Official Telephone Number",

    # Dates
    "Certification Date",
    "Last Update Date",
]


writer = None

for chunk in pd.read_csv(csv_path, chunksize=chunk_size, usecols=columns_to_keep, low_memory=False):
    table = pa.Table.from_pandas(chunk)
    if writer is None:
        writer = pq.ParquetWriter(parquet_path, table.schema)
    writer.write_table(table)

if writer:
    writer.close()

print(f" Parquet file saved: {parquet_path}")
```

--filter the trimmed parquet

```python
import polars as pl

# Load only needed columns
df = pl.read_parquet(parquet_path, columns=[
    "NPI",
    "Entity Type Code",
    "Provider Organization Name (Legal Business Name)",
    "Provider Business Mailing Address State Name",
    "Healthcare Provider Taxonomy Group_1"
])

# Filter orgs in FL with 'radiology' taxonomy
filtered = df.filter(
    (pl.col("Entity Type Code") == 2) &
    (pl.col("Provider Business Mailing Address State Name") == "FL")&
    (pl.col("Healthcare Provider Taxonomy Group_1").str.to_lowercase().str.contains("Radiology"))
)


# Save to Parquet or CSV
filtered.write_parquet(r"C:\Users\ChristopherCato\Downloads\npi_FL_radiology.parquet")
# Or: filtered.write_csv(r"...")

print(f" Filtered result: {filtered.shape[0]} rows")
```
